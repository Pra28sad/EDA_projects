```python
## generally price is in integer but here it is in object

## so we need to convert into integer data

df["discount_percentage"]
```

```python

def convert_into_int(data):
    if "%" in data:
        data = data.replace("%","")
    return int(data)

df["discount_percentage"] = df["discount_percentage"].apply(convert_into_int)
```

```python
df["discount_percentage"] ## we have successfully converted into int
```

#  1) What is the average rating for each product category?

```python
df["primary_category"]=df["category"].apply(lambda x:x.split("|")[0])
```

```python
avg = df.groupby("primary_category")["rating"].mean().sort_values(ascending=False)
```

# 3)What is the distribution of discounted prices vs. actual prices?


```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["actual_price"],y=df["discounted_price"])
plt.title("Distribution of discounted Prices vs Actual Prices")
plt.xlabel("Actual Prices")
plt.ylabel("Discounted Prices")
plt.show()
```

#  How does the average discount percentage vary across categories?

```python
avg_discount = df.groupby(["primary_category"])["discount_percentage"].mean()
```

```python
plt.figure(figsize=(10,6))
plt.pie(avg_discount.sort_values(ascending=False),labels=df["primary_category"].unique(),autopct="%.0f%%")
plt.show()
```

```python
avg_discount.sort_values(ascending=False)
```
