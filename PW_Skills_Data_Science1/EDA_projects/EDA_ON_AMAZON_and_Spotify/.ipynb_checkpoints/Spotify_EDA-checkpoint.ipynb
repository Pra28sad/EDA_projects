###  2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.

```python
plt.figure(figsize=(10,6))
sns.histplot(x=df['Popularity'],color="red")
plt.title("Distribution among the tracks and popularity")
plt.show()

```

### 3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x="Popularity",y="Duration (ms)")
plt.title("Relation between popularity and duration of tracks")
plt.show()
```

###  4.Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.


```python
highest_nof_tracks  = df["Artist"].value_counts().reset_index()
```

```python
plt.figure(figsize=(10,30))
sns.countplot(data = df,y = "Artist" )
plt.show()
```

# 5.What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.

```python
data_sorted = df["Popularity"].sort_values(ascending=True)
df[["Artist","Track Name","Popularity"]].sort_values(by="Popularity")[:5]
```

###  6.Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist.

```python
top_5_avg = df.groupby(["Artist"])["Popularity"].mean().sort_values(ascending = False).reset_index()
```

```python
top_5_avg.head()
```

### 7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist

```python
df[["Artist","Track Name","Popularity"]].sort_values(by="Popularity",ascending=False)[:5]
```

###  8.Visualize relationships between multiple numerical variables simultaneously using a pair plot.

```python
sns.pairplot(df)
# plt.title("Relation between multipe numerical  variables")
plt.show()
```

### 9.Does the duration of tracks vary significantly across different artists?  Explore this visually using a box plot or violin plot.

```python
plt.figure(figsize=(15,20))
sns.boxplot(data=df,y = "Artist",x = "Duration (ms)")
plt.show()
```

### 10.How does the distribution of track popularity vary for different artists?Visualize this using a swarm plot or a violin plot

```python
plt.figure(figsize=(12,12))
sns.violinplot(data=df,x='Popularity',y='Artist')
```
